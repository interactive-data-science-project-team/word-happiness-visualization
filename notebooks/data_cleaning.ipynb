```python
import pandas as pd
```

```python
data = pd.read_excel('../input/WHR2018Chapter2OnlineData.xls')
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data = data.rename(index=str, columns={"Life Ladder": "Happiness Score", 
                                       "Standard deviation of ladder by country-year": "Standard deviation of happiness score by country-year",
                                       "Standard deviation/Mean of ladder by country-year": "Standard deviation/Mean of happiness score by country-year"})
```

```python
data.count()
```

```python
na_count_df1 = pd.DataFrame(data.isna().sum())
na_count_df1.columns = ['na_count_before']
na_count_df1['na_percentage_before'] = na_count_df1['na_count_before'] / data.shape[0]

na_count_df1.head()
```

```python
report_by_year = pd.DataFrame(data.groupby('year')['country'].count())
report_by_year.columns = ['num_countries']
report_by_year.head()
```

```python
report_by_year.to_csv('../app/data/num_report_by_year.csv', index_label='year')
```

```python
data.groupby('country').head()
```

```python
cleaned_data = data.groupby('country', as_index=False).fillna(method='ffill')
cleaned_data = cleaned_data.groupby('country', as_index=False).fillna(method='bfill')
cleaned_data.head()
```

```python
na_count_df2 = pd.DataFrame(cleaned_data.isna().sum())
na_count_df2.columns = ['na_count_after']
na_count_df2['na_percentage_after'] = na_count_df2['na_count_after'] / cleaned_data.shape[0]
```

```python
na_count_df = na_count_df1.merge(na_count_df2, left_index=True, right_index=True)
na_count_df.head()
```

```python
na_count_df.to_csv('../app/data/na_counts.csv', index_label='attribute')
```
