```python
happiest2017 = df.loc[df['year'] == 2017].sort_values('Life Ladder', ascending=False)[0:5]['country'].values
mask = df.country.map(lambda c: c in happiest2017)
```

```python
plt.figure(figsize=(12, 6))

sns.lineplot(x='year', y='Life Ladder', hue='country', data=df[mask])
plt.ylabel('Happiness Score')

plt.show()
```

```python
plt.figure(figsize=(12, 12))

for i, x in enumerate(['Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity']):
    plt.subplot(2, 2, i + 1)
    sns.regplot(x=x, y='Life Ladder', data=df.loc[df['year'] == 2017])
    plt.ylabel('Happiness Score')

plt.show()
```
